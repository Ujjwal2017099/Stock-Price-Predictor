```python
db_url = "GOOG.csv"
```

```python
import pandas as pd
```

```python
db = pd.read_csv(db_url)
```

```python
db
```

```python
db.info()
```

```python
df = pd.DataFrame(db)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.set_index('Date',inplace=True)
```

```python
df['Change'] = df['Open']-df['Close']
df
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
inputs = df[['Open','Volume']]
targets = df['Close']
```

```python
input_train , input_test , output_train , output_test = train_test_split(inputs,targets,test_size=0.2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
poly_features = PolynomialFeatures(degree=2)
input_train_poly = poly_features.fit_transform(input_train)
input_test_poly = poly_features.fit_transform(input_test)
```

```python
model = LinearRegression()
model.fit(input_train_poly,output_train)
```

```python
test_predict = model.predict(input_test_poly)
# test_predict
```

```python
import numpy as np
```

```python
def rmse(target,prediction):
    return np.sqrt(np.mean(np.square(target-prediction)))
```

```python
rmse(output_test,test_predict)
```
